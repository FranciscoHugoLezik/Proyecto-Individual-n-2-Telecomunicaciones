## Extracción del archivo

## Selección y visualización de los datasets a analizar mediante EDA

Se seleccionan los siguientes datasets:

```python
seleccionados = [
    'Accesos por rangos', 
    'Accesos Por Tecnología', 
    'Penetracion-hogares'
    ]
```

Se explica porque se seleccionaron:

```python
explicación = {
    'Accesos por rangos': 'Explicación: quiero saber si la tendencia es hacia mas de 30 Mbps o un rango intermedio', 
    'Accesos Por Tecnología': 'Explicación: quiero saber que tecnologia es la que mas crece.', 
    'Penetracion-hogares': 'Explicación: quiero saber que provincias tienen mas acceso a internet cada 100 hogares.'
}
```

Se visualizan los datasets y las explicaciones:

```python
for dataset in seleccionados:
    print()
    display(
        dataset, 
        explicación[dataset], 
        internet_diccionario[dataset]
        )
    print()
```

### Exportar el dataset 'Accesos por rangos'

Se lo va a exportar a la carpeta data.

```python
ruta_a_exportar = os.path.join(
    ruta_del_proyecto, 
    'data', 
    'accesos_por_rangos.parquet'
    )

ruta_a_exportar
```

Se lo exporta.

```python
internet_diccionario['Accesos por rangos'].to_parquet(
    ruta_a_exportar, engine='fastparquet')
```

### Exportar el dataset 'Accesos Por Tecnología'

Se lo va a exportar a la carpeta data.

```python
ruta_a_exportar = os.path.join(
    ruta_del_proyecto, 
    'data', 
    'accesos_por_tecnologia.parquet'
    )

ruta_a_exportar
```

Se va a hacer unas transformaciones para que se pueda exportar.

```python
accesos_por_tecnologia = internet_diccionario['Accesos Por Tecnología']
```

```python
accesos_por_tecnologia
```

Se eliminan las filas con el valor '2019 *'

```python
accesos_por_tecnologia[accesos_por_tecnologia['Año'] == '2019 *']
```

```python
indices_a_eliminar = accesos_por_tecnologia[
    accesos_por_tecnologia['Año'] == '2019 *'].index

accesos_por_tecnologia.drop(indices_a_eliminar, inplace=True)
```

Se elimina la fila con 1008 que tiene un '*'.

```python
accesos_por_tecnologia.drop(1008, inplace=True)
```

```python
accesos_por_tecnologia
```

```python
accesos_por_tecnologia.reset_index(inplace=True, drop=True)

accesos_por_tecnologia
```

Se lo exporta.

```python
accesos_por_tecnologia.to_parquet(
    ruta_a_exportar, engine='fastparquet')
```
