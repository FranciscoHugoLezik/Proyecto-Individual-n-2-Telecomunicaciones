## Extracción del dataset

```python
ruta_actual = os.getcwd()

ruta_actual
```

```python
ruta_del_proyecto = os.path.dirname(
    ruta_actual
    )

ruta_del_proyecto
```

```python
ruta_del_dataset = os.path.join(
    ruta_del_proyecto, 
    'data', 
    'Internet.xlsx'
    )

ruta_del_dataset
```

```python
internet_diccionario = pd.read_excel(ruta_del_dataset, sheet_name=None)
```

```python
for key in internet_diccionario.keys():
    display(key, internet_diccionario[key])
    display()
```

```python
internet_diccionario['Penetracion-totales']
```

```python
penetracion_totales = internet_diccionario['Penetracion-totales']
```

```python
penetracion_totales
```

```python
penetracion_totales.info()
```

```python
plt.figure(figsize=(20,10))

sns.lineplot(
    data=penetracion_totales, 
    x='Año', 
    y='Accesos por cada 100 hogares'
    )

plt.show()
```

```python
plt.figure(figsize=(20,10))

sns.lineplot(
    data=penetracion_totales, 
    x='Año', 
    y='Accesos por cada 100 hab' 
    )

plt.show()
```

```python
totales_accesos_por_tecnología = internet_diccionario['Totales Accesos Por Tecnología']

totales_accesos_por_tecnología
```

```python
df_long = pd.melt(totales_accesos_por_tecnología, id_vars=['Año'], value_vars=['ADSL', 'Cablemodem', 'Fibra óptica', 'Wireless', 'Otros'], 
                  var_name='Tecnología', value_name='Suscriptores')
```

```python
plt.figure(figsize=(14, 8))
sns.lineplot(data=df_long, x='Año', y='Suscriptores', hue='Tecnología', marker='o')
plt.title('Comparación de Tecnologías de Internet a lo Largo de los Años')
plt.xlabel('Año')
plt.ylabel('Cantidad de Suscriptores')
plt.legend(title='Tecnología')
plt.grid(True)
plt.show()
```
